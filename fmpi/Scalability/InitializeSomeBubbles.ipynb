```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
from scipy.integrate import odeint
```

```python
X0 = Y0 = 0.;
L0 = 2.;
Nb = 20
n = 0;

_xc = []
_yc = []
_rc = []

rng1 = np.random.RandomState(10)  # deterministic random data
rng2 = np.random.RandomState(11)  # deterministic random data
rng3 = np.random.RandomState(12)  # deterministic random data
_x = rng1.normal(size=1000)
_y = rng2.normal(size=1000)
_r = rng3.normal(size=1000)

eps = 0.01*L0;

for i in range(1000):
    x = X0 + 0.5*L0 + 0.5*L0*_x[i];
    y = Y0 + 0.25*L0 + 0.25*L0*_y[i];
    r = 0.025 * L0 * (1 + _r[i]) ;
    if ( r < 0.02*L0 or r > 0.05*L0):
        continue
    if ( x-r < X0+eps or x+r > X0+L0-eps):
        continue;
    if ( y-r < Y0+eps or y+r > Y0+0.5*L0):
        continue;
    valid = True;
    for j in range(n):
        if ( ((x - _xc[j])**2 + (y - _yc[j])**2)**0.5 < r+_rc[j] + eps):
            valid = False;
            break;
    if valid == True :
        _xc.append(x); _yc.append(y); _rc.append(r);
        n=n+1;
    if n == Nb:
        break;
    
print(_xc);
print(_yc);
print(_rc);
```

```python
fig, ax = plt.subplots() 

circles = [];

for i in range(len(_xc)):
        ax.add_patch(plt.Circle((_xc[i], _yc[i]), _rc[i], color='r'))

#     circle = plt.Circle((_xc[j], _yc[j]), _rc[j], color='r')
#     circles.append(circle);
#     plt.gca().add_patch(circle1)
#     ax.add_patch(circles[i])

# c = plt.Circle((_xc[10], _yc[10]), _rc[10], color='r')
# ax.add_patch(plt.Circle((_xc[10], _yc[10]), _rc[10], color='r'))
# ax.add_patch(plt.Circle((_xc[12], _yc[12]), _rc[12], color='r'))


plt.xlim([0, 2])
plt.ylim([0, 2])
plt.show()

# fig.savefig('plotcircles.png')
```
