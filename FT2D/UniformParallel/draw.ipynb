```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
```

```python
text = ""
```

```python
text = text + "%Bubble 0\n";
Circle0 = [4.6, 2.0, 0.8]
Np = int(np.floor(2*Circle0[2]*np.pi/0.5))
theta = 2*np.pi/Np

for i in range(Np):
    th = theta*(i+0.5);
    xm = Circle0[0] + Circle0[2]*np.cos(th)
    ym = Circle0[1] + Circle0[2]*np.sin(th)
    pid = int(np.floor(xm/5.)*2 + np.floor(ym/5.))
    text = text +  "  \draw[elems, pid" + str(pid)+ "]"
    th = theta*i;
    x = Circle0[0] + Circle0[2]*np.cos(th)
    y = Circle0[1] + Circle0[2]*np.sin(th)
    strx = '%.5f' % x
    stry = '%.5f' % y
    text = text +  "(" + strx + "," + stry + ")" + "--"
    th = theta*(i+1);
    x = Circle0[0] + Circle0[2]*np.cos(th)
    y = Circle0[1] + Circle0[2]*np.sin(th)
    strx = '%.5f' % x
    stry = '%.5f' % y
    text = text +  "(" + strx + "," + stry + ");" + "\n"

for i in range(Np):
    th = theta*(i);
    x = Circle0[0] + Circle0[2]*np.cos(th)
    y = Circle0[1] + Circle0[2]*np.sin(th)
    pid = int(np.floor(x/5.)*2 + np.floor(y/5.))
    text = text +  "  \draw[pid" + str(pid)+ "]"
    strx = '%.5f' % x
    stry = '%.5f' % y
    text = text +  "plot [pmarker] coordinates {(" + strx + "," + stry + ")" + "};" + "\n";
    
print(text)
```
