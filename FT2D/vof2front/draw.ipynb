```python
f = open('front.dat', 'r')
data = np.genfromtxt(f, delimiter=',');
f.close()
l, = np.shape(data);
l = (int) (l/7)
edges = np.reshape(data, (l,-1))

%matplotlib notebook

ax = plt.gca(projection="3d")

colors = ['r','b','g','m','c','y','g','k','r'] 
for i in range(l):
    edge = np.copy(edges[i]);
    xa,xb,ya,yb,za,zc, c = edge
    c = (int) (c)
    ax.plot([xa,xb],[ya,yb],[za,zc], colors[c], linewidth = 1.)


xlim = [0,1.]
ylim = [0,1.]
zlim = [0,1.]
ax.set_xlim(xlim);ax.set_ylim(ylim);ax.set_zlim(zlim);

plt.show()     
```
