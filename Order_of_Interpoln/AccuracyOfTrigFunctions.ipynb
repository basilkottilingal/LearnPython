```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
from scipy.optimize import minimize
import math
```

```python
def sin(th):
    sth = 0.
    for i in range(20):
        m = 1.;
        fact = 1.;
        for j in range(2*i+1):
            m *= th
            fact *= (j+1) 
        sth += np.power(-1,i)*m/fact;
    return(sth);
```

```python
err = [];
theta = 2*np.pi*np.arange(0, 1.0000001, 0.01);
theta2 = 2*np.pi*np.arange(1.000000,-0.00001, -0.01);
```

```python
fig, ax1 = plt.subplots()
ax1.set_xlabel("theta")
ax1.set_ylabel("sin(theta) - np.sin(theta)")
ax1.semilogy(theta, abs(sin(theta)-np.sin(theta)), linewidth = 0.7, label = "err")
ax1.semilogy(theta, abs(-sin(theta2)-np.sin(theta)), linewidth = 0.7, label = "err")

ax1.semilogy(theta, abs(sin(theta)+np.sin(theta2)), linewidth = 0.7, label = "err")
ax1.semilogy(theta, abs(-sin(theta2)+np.sin(theta2)), linewidth = 0.7, label = "err")


ax1.legend(loc = "lower right")

fig.tight_layout()  # otherwise the right y-label is slightly clipped

plt.show()
```
