# Linearisation of R-P equation in bulk
A bubble $R_0$ initially at equilibrium with the ambient pressure $p_{\infty,0}$ has a
equilibrium bubble pressure $p_{B,0} = p_{G,0} + p_V (T_\infty)$. The bubble response
in the limit of $M\rightarrow 0$ is given by R-P equation
\begin{equation}
  R\ddot{R} + \frac{3}{2} \dot{R}^2 + \frac{4\mu}{\rho_L} \frac{\dot{R}}{R} + \frac{2\sigma}{\rho_L R} =
  \frac{1}{\rho_L} \left(p_B(t) - p_\infty (t)\right)
\end{equation}
In the absence of vapor pressure and the gas inside bubble follows polytropic gas law,
we have
\begin{equation}
  p_B (t) = p_{G,0} \left( \frac{R_0}{R} \right)^{3k}
\end{equation}
If the ambient pressure is perturbed as
\begin{equation}
  p_\infty (t) = p_{\infty,0} + p_A ~R e \{ e^{j\omega t}\}
\end{equation}
with $p_A>0, ~p_A \ll p_{\infty,0}$
produces a linearised pertubation in the bubble radius from its equilibrium radius of the form
\begin{equation}
  R (t) = R_0 \left(1 - \epsilon_0 ~R e \{ e^{j(\omega t +\phi)} \} \right)
\end{equation}
with $\epsilon_0>0, ~\epsilon_0 \ll 1$.

\begin{eqnarray*}
  R\ddot{R} &=& \omega^2 R_0^2 (1 - \epsilon_0 { e^{j(\omega t +\phi)} })
                             \epsilon_0 { e^{j(\omega t +\phi)} } \\
  \frac{3}{2} \dot{R}^2 &=& \frac{3}{2} R_0^2 \omega^2 \epsilon_0^2 { e^{2j(\omega t +\phi)} }  \\
  \frac{4\mu_L}{\rho_L} \frac{\dot{R}}{R} &=& -j\omega\epsilon_0 ~{ e^{j(\omega t +\phi)} }
                                  (1+\epsilon_0 ~{ e^{j(\omega t +\phi)} })\frac{4\mu_L}{\rho_L} \\
  \frac{2\sigma}{\rho_L R} &=& \frac{2\sigma}{\rho_L R_0}
                   (1+\epsilon_0 { e^{j(\omega t +\phi)} } + \epsilon_0^2 ~{ e^{2j(\omega t +\phi)} } )\\
  p_B(t) &=& p_{G,0} (1+3k\epsilon_0 ~{ e^{j(\omega t +\phi)} }
                    +\frac{3k(3k-1)}{2}\epsilon_0^2 ~{ e^{2j(\omega t +\phi)} }) \\
  p_\infty(t) &=& p_{\infty,0} + p_A  e^{j \omega t }
\end{eqnarray*}
which gives
\begin{eqnarray}
  \epsilon_0^2 ~{ e^{2j(\omega t +\phi)} } \left(
    -\omega^2 R_0^2  + \frac{3}{2} R_0^2 -j\omega \frac{4\mu_L}{\rho_L} + \frac{2\sigma}{\rho_L R_0}
    - \frac{p_{G,0}}{\rho_L} \frac{3k(3k-1)}{2}
  \right) + \nonumber\\
  \epsilon_0 ~{ e^{j(\omega t +\phi)} } \left(
    \omega^2 R_0^2 - j\omega \frac{4\mu_L}{\rho_L} + \frac{2\sigma}{\rho_L R_0} - \frac{3kp_{G,0}}{\rho_L}
  \right)
  =
  \frac{1}{\rho_L} \left( p_{G,0} - \frac{2\sigma}{ R_0} - p_{\infty,0}  \right) - \frac{p_A}{\rho_L} e^{j\omega t}
\end{eqnarray}

# Natural Frequency inside rigid  tube [Oguz and Prosperetti]

A bubble centered at the axis $(0,0)$ inside a rigid tube of length $L$ and radius $R_{v}$ and a negligible thickness $\Gamma_t = \{ (r,z)~|~ r=R, ~-L/2 \leq z \leq L/2\}$. The bubble of initial radius $R_0$ is driven by acoustic cavitation which makes the bubble interface, $\Gamma$, oscillates

## Natural frequency of bubble oscillating inside tube

The bubble pressure inside the bubble and the liquid pressure just outside the interface can be found by stress balance.
(neglecting viscosity)
$$
    p_B(\mathbf{x},t) = p_L (\mathbf{x},t) + \sigma \kappa(\mathbf{x},t) \quad \forall \mathbf{x} \in \Gamma 
$$
    Using the unsteady incompressible potential flow equation in liquid $\frac{p_L}{\rho} = \frac{p_\infty}{\rho} - \frac{\partial \phi } {\partial t} - \frac{1}{2} \left( \nabla \phi \right)^2$. Neglecting $ \frac{1}{2} \left( \nabla \phi \right)^2$ w.r.t $\frac{\partial \phi } {\partial t}$ we get
$$
    p_{B}=-\rho \frac{\partial\phi}{\partial t}  + p_{\infty}+\sigma \kappa 
$$


Averaging on the onterface
$$
    p_{B}=-\rho\langle \frac{d\phi}{d t} \rangle +p_{\infty}+\sigma\langle \kappa \rangle,
$$
Differentiating
$$
    \frac{d}{d t}p_{B}=-\rho\langle \frac{d^2\phi}{d t^2} \rangle + \sigma\frac{d}{d t} \langle \kappa \rangle,
$$
    


$$
    \frac{d p_{B} }{d t} = \frac{d p_{B} }{d V} \frac{d V }{d t} 
$$
$${\frac{d V}{d t}}=\int_{\cal S}\mathbf{u}\cdot\mathbf{n}\ d{\cal S}\equiv{\cal S}\biggl<{\frac{\partial\phi}{\partial n}}\biggr>
$$

$$\phi =\sum_{n=0}^{\infty}\;A_{n}\,\frac{a^{n+1}}{\rho^{n+1}}\,P_{n}(\cos\,\theta)+\sum_{n=1}^{\infty}\;B_{n}I_{0}\biggl(\frac{n\pi r}{L}\biggr)\sin\frac{n\pi z}{L} +\,C_{0}\,\frac{z}{L}+D_{0}\,\frac{L-z}{L}+\sum_{n=1}^{\infty}\left[\,C_{n}\,\frac{\sinh\alpha_{n}(z/R)}{\sinh\,\alpha_{n}(L/R)}\,\right.  + D_{n}\,\frac{\sinh\,\alpha_{n}(L-z)/R}{\sinh\,\alpha_{n}(L/R)}\Big|J_{0}\Big(\,\alpha_{n}\,\frac{r}{R}\Big)
$$

## Constants $A_n, B_n, C_n, D_n$
using the velocity contiion at the tube wall  $ \frac{\partial}{\partial r} \phi = 0$
$$\frac{\partial}{\partial r} \left[ \sum_{n=0}^{\infty}\;A_{n}\,\frac{a^{n+1}}{\rho^{n+1}}\,P_{n}(\cos\,\theta)+\sum_{n=1}^{\infty}\;B_{n}I_{0}\biggl(\frac{n\pi r}{L}\biggr)\sin\frac{n\pi z}{L} \right] = 0
$$
Outflow condition at tube exit gives $\frac{\partial}{\partial r} \phi = 0$
$$+\sum_{n=1}^{\infty}\;B_{n}\;\frac{n\,\pi}{L}\;I_{1}\bigg(\frac{n\,\pi R}{L}\bigg)\mathrm{sin}\;\frac{n\,\pi z}{L}=0.
$$
$$\sum_{n=0}^{\infty}\;A_{n}\;\frac{a^{n+1}}{\rho^{n+1}}\;P_{n}(\cos\;\theta)+\sum_{n=1}^{\infty}\;D_{n}J_{0}\biggl(\alpha_{n}\;\frac{r}{R}\biggr)+D_{0}=0,

$$\sum_{n=0}^{\infty}\;A_{n}\,\frac{a^{n+1}}{\rho^{n+1}}\,P_{n}(\cos\,\theta)\sum_{n=1}^{\infty}\;C_{n}J_{0}\bigg(\,\alpha_{n}\,\frac{r}{R}\bigg)+C_{0}=0.
$$
$$\frac{B_{n}}{A_{0}}=\frac{2}{n\pi}\,\frac{1}{I_{1}(n\pi R/L)}\;\int_{0}^{L}\frac{R}{[R^{2}+(z-d)^{2}]^{3/2}}\,\sin\biggl(\frac{n\pi z}{L}\biggr)d z

```python
from scipy import special
import math
## Constants in velocity potential functions
nterms = 5;
nsteps = 1000; 
def phi_params(L, d):
    ## All the dimensions are assumed to be normalized with R_v
    ## All constants A[i], B[i], C[i], D[i] are normalized with A[0]
    A = np.zeros(nterms);
    A[0] = 1.;
    B = np.zeros(nterms);
    C = np.zeros(nterms);
    D = np.zeros(nterms);
    alpha = special.jn_zeros(1, nterms)
    for n in range(1):
        C[n] = 0;
        for j in range(nsteps):
            dr =  1/nsteps;
            r = j*dr;
            #trapezoidal summation
            C[n] -=  2 * r * dr / ((r**2 +  (L-d)**2)**0.5) 
            r += dr
            C[n] -=  2 * r * dr / ((r**2 +  (L-d)**2)**0.5)
        D[n] = 0;
        for j in range(nsteps):
            dr =  1/nsteps;
            r = j*dr;
            #trapezoidal summation
            D[n] -=  2 * r * dr / ((r**2 +  d**2)**0.5) 
            r += dr
            D[n] -=  2 * r * dr / ((r**2 +  d**2)**0.5) 
        C[n] /= 2.
        D[n] /= 2.
    for _i in range(nterms-1):
        n = _i + 1;
        B[n] = 0;
        for j in range(nsteps):
            dz =  L/nsteps;
            z = j*dz;
            #trapezoidal summation
            B[n] +=  (2/(n*np.pi)) * (1/special.i1(n*np.pi/L) ) * np.sin(n*np.pi*z/L) * dz/((1 + (z-d)**2)**1.5)
            z += dz
            B[n] +=  (2/(n*np.pi)) * (1/special.i1(n*np.pi/L) ) * np.sin(n*np.pi*z/L) * dz/((1 + (z-d)**2)**1.5)
        C[n] = 0;
        for j in range(nsteps):
            dr =  1/nsteps;
            r = j*dr;
            #trapezoidal summation
            C[n] -=  (2/(special.j0(alpha[n]))**2) * r * special.j0(alpha[n]*r) * dr / ((r**2 +  (L-d)**2)**0.5) 
            r += dr
            C[n] -=  (2/(special.j0(alpha[n]))**2) * r * special.j0(alpha[n]*r) * dr / ((r**2 +  (L-d)**2)**0.5) 
        D[n] = 0;
        for j in range(nsteps):
            dr =  1/nsteps;
            r = j*dr;
            #trapezoidal summation
            D[n] -=  (2/(special.j0(alpha[n]))**2) * r * special.j0(alpha[n]*r) * dr / ((r**2 +  d**2)**0.5) 
            r += dr
            D[n] -=  (2/(special.j0(alpha[n]))**2) * r * special.j0(alpha[n]*r) * dr / ((r**2 +  d**2)**0.5) 
        B[n] /= 2.;
        C[n] /= 2.
        D[n] /= 2.
    return (A,  B, C, D, alpha);

   
```

### Velocity $\mathbf{u} = \frac{\partial \phi}{\partial r} \mathbf{e}_r + \frac{\partial \phi}{\partial z} \mathbf{e}_z$

```python

def velocity(params, r, z, R_0, L, d):
    A, B, C, D, alpha = params
    phi = 0.;
    ur = 0.;
    uz = 0.;
    # limited to 0 (for A[n])
    _rho = np.sqrt(r**2 + (z-d)**2);
    _p, _dp = special.lpn(0, (z-d)/_rho );
    for n in range(1):
        common = A[n] * ((R_0/_rho)** (n+1) )
        phi += common * _p[n];
        ur -=   common * ( _p[n] * (n+1) * r /( _rho**2 ) + _dp[n] * r * (z-d) / (_rho**3) );
        uz -=   common * ( _p[n] * (n+1) * (z-d) /( _rho**2 ) - _dp[n] * r * r / (_rho**3) );
    for _i in range(nterms-1):
        n = _i+1;
        phi += B[n] * np.sin(n*np.pi*z/L) * special.i0(n*np.pi*r/L)
        ur += B[n] * np.sin(n*np.pi*z/L) * special.i1(n*np.pi*r/L) *(n*np.pi/L);
        uz += B[n] * np.cos(n*np.pi*z/L) * special.i0(n*np.pi*r/L) *(n*np.pi/L);
    phi += C[0]*z/L + D[0]*(L-z)/L
    ur += 0;
    uz += (C[0]-D[0])/L
    for _i in range(nterms-1):
        n = _i+1;
        phi += (C[n]*math.sinh(alpha[n]*z)/math.sinh(alpha[n]*L) + D[n]*math.sinh(alpha[n]*(L-z))/math.sinh(alpha[n]*L)) * special.j0(alpha[n]*r)
        ur -=  (C[n]*math.sinh(alpha[n]*z)/math.sinh(alpha[n]*L) + D[n]*math.sinh(alpha[n]*(L-z))/math.sinh(alpha[n]*L)) * alpha[n] *special.j1(alpha[n]*r)
        uz +=  (C[n]*math.cosh(alpha[n]*z)/math.sinh(alpha[n]*L) - D[n]*math.cosh(alpha[n]*(L-z))/math.sinh(alpha[n]*L)) * alpha[n] *special.j0(alpha[n]*r)
    return (ur, uz, phi)


```

###  Frequency  ratio $\omega^{2} /\omega_0^2$
$$
    \omega^{2}=-\frac{1}{\rho} \left(S \,\frac{d p_{B}}{d V}+\frac{2\,\sigma}{a^{2}}\right)\frac{-1}{\langle \phi \rangle}\left<\frac{\partial\phi}{\partial n}\right> 
$$
Bubble in the bulk case $\phi = -\dot{R} R^2/r$ which gives $\mathbf{u}\cdot \mathbf{n} \bigg|_{r=R} = \frac{\partial \phi}{\partial n} = \dot{R}$ and $\left< \frac{1}{2} \right> $

$$
    \frac{\omega^{2}}{\omega_0^{2}}= - \frac{R_0}{\langle \phi \rangle}\left<\frac{\partial\phi}{\partial n}\right> 
$$


```python
def phi_average(params, R_0, L, d):
    #params etc are set
    A, B, C, D, alpha = params
    phi_av = 0.;
    dtheta =  2*np.pi/nsteps;
    for i in range(nsteps):
        #left end of trapezium summ
        theta = i*dtheta;
        z = d + R_0 * np.cos(dtheta); 
        r = R_0 * np.sin(dtheta);
        ur, uz, phi = velocity(params, r, z, R_0, L, d);
        phi_av += phi * (2*R_0**2) * dtheta
        #right end of trapezium summ
        theta += dtheta;
        z = d + R_0 * np.cos(dtheta); 
        r = R_0 * np.sin(dtheta);
        ur, uz, phi = velocity(params, r, z, R_0, L, d);
        phi_av += phi * (2*R_0**2) * dtheta 
    phi_av /= 2.; # Trapezium summation
    phi_av /= 4.*np.pi*R_0**2
    return phi_av;

def dphidn_average(params, R_0, L, d):
    #params etc are set
    A, B, C, D, alpha = params
    dphidn_av = 0.;
    dtheta =  2*np.pi/nsteps;
    for i in range(nsteps):
        theta = i*dtheta;
        z = d + R_0 * np.cos(dtheta); 
        r = R_0 * np.sin(dtheta);
        ur, uz, phi = velocity(params, r, z, R_0, L, d);
        dphidn_av += dtheta* (2*R_0**2) * (uz*np.cos(theta) + ur*np.sin(theta))
        theta += dtheta;
        z = d + R_0 * np.cos(dtheta); 
        r = R_0 * np.sin(dtheta);
        ur, uz, phi = velocity(params, r, z, R_0, L, d);
        dphidn_av += dtheta* (2*R_0**2) * (uz*np.cos(theta) + ur*np.sin(theta))
    dphidn_av /= 2.; # Trapezium summation
    dphidn_av /= 4.*np.pi*R_0**2
    return dphidn_av;
```

# Linearised  R-P equation
For an acoustic excitation 
    $$ p_A ~Re \{ e^{j ( \omega t) } \} $$ 
the bubble responds as 
    $$ R(t) = R_0  -  R_0 \epsilon_0 Re \left\{ e^{j ( \omega t + \phi)} \right\}$$
where $\epsilon_0$ and $\phi$ are solved using the response equation 
$$
\omega^2  - j\omega \frac{4\mu_L}{\rho_L R_0^2} - \omega_0^2 = - \frac{p_A}{\rho_L R_0^2 \epsilon_0} e^{-j\phi}
$$
where
$$
\omega_0^2 = \frac{1}{\rho_L R_0^2} \left(  {3kp_{G,0}} - \frac{2\sigma}{R_0} \right)
$$

$$
\phi = - arctan {
            \bigg( \frac{ \omega^2 - \omega_0^2}{ ( \omega \frac{4\mu_L}{\rho_L R_0^2}  )} \bigg)
                 } - \frac{\pi}{2}
$$

$$
\epsilon_0 = \frac{p_A}{\rho_L R_0^2}
  \frac{1}{\sqrt{\left(  (\omega^2 - \omega_0^2)^2 + (\omega \frac{4\mu_L}{\rho_L R_0^2} )^2        \right)}}
$$

## Non Dimensional numbers

(1) Acoustic pressure to ambient pressure ratio
$$
    p^* = \frac{p_A}{p_{\infty,0}}
$$
(2) Weber Number
$$
    w^* = \frac{p_{\infty,0} R_0}{2\sigma}
$$
(3) 
$$
 l^{*} = \frac{p_{\infty,0}}{\rho_L R_0^2 \omega^2}
$$
(4) Ratio of natural frequency to driving frequency ratio
$$
    \omega^* = \frac{\omega_0}{\omega} = \sqrt{l^* \left( 3k (1+\frac{1}{w^*})  - \frac{1}{w^*}\right) }
$$
(5) Viscous Damping coefficient
$$
    \zeta^* = \frac{\beta_{\mu}}{\omega} = \frac{2\mu_L}{\rho_L R_0^2 \omega}
$$
Don't confuse with
$$
    \zeta = \frac{\beta_{\mu}}{\omega_0} = \frac{2\mu_L}{\rho_L R_0^2 \omega_0} = \frac{\zeta^*}{\omega^*}
$$
## Bubble response
which gives for an external pressure field with an acoustic pressure pertubration
$$
    p_{\infty}(t) = p_{\infty,0} \left( 1 + p^* Re \{ e^{j\omega t}\} \right)
$$
The bubble responds as
$$
R(t) = R_0  -  R_0 \epsilon_0
    ~Re
      \left\{ e^{j ( \omega t + \phi)} \right\}
$$
where 
$$
\phi = - arctan {
            \bigg( \frac{ 1 - {\omega^*}^2 }{ ( 2\zeta^* \omega^* )} \bigg)
                 } - \frac{\pi}{2}
$$
and 
$$
\epsilon_0 = 
  \frac{p^* l^*}{\sqrt{\left(  {(1 - {\omega^*}^2)}^2 + { ( 2\zeta^* \omega^* )}^2        \right)}}
$$

### Transient Effect
$$
R(t) = R_0  
    - R_0 \epsilon_0 ~ Re \left\{ e^{j ( \omega t + \phi)} \right\} 
    + R_0 \epsilon_0 e^{\zeta \omega_0 t} ~ Re \left\{ e^{j ( \omega_d t + \phi)} \right\}
$$

$$
R^*(t^*) = 1  
    -  \epsilon_0 ~ Re \left\{ e^{j ( t^* + \phi)} \right\} 
    +  \epsilon_0 e^{-\zeta^* t^*} ~ Re \left\{ e^{j \left(t^*  \sqrt{{\omega^*}^2 - {\zeta^*}^2 } + \phi \right) } \right\}
$$
which gives $R^*(t^* = 0) =  1$

```python
#NOn dimensional numbers
def parameters(omegastar, R_0, k, pstar):
    #weber number
    wstar = p_atm/(2*sigma/R_0)
    #Natural Frequency
    omega0 = np.sqrt(p_atm/(rho*R_0**2))*np.sqrt( 3*k*(1+1/wstar) - 1/wstar );
    omega = omega0/omegastar
    lstar = p_atm/(rho* R_0**2 * omega**2);
    zetastar = 2*mu/(rho* R_0**2 * omega);
    return [pstar, wstar, lstar, omegastar, zetastar, omega0]    
```

```python
#Amplitude and phase
def cosineparameters(pstar, wstar, lstar, omegastar, zetastar):
    phi = -np.arctan2(1-omegastar**2, 2*zetastar)-0.5*np.pi
    eps = pstar*lstar/np.sqrt( ( 1-omegastar**2 )**2 +  ( 2*zetastar )**2);
    return [phi, eps]
```

```python
def cosine(A, PHI, OMEGA, T):
    return A*np.cos(OMEGA*T+PHI)
```

```python
def linear_response(tstar_space):
    #tstar = omega t;
    #assume non-dim parameters are set
    phi,eps = cosineparameters(pstar, wstar, lstar, omegastar, zetastar);
    _1 = np.ones(len(tstar_space));
    Rstar_space = _1 - cosine(eps, phi, 1, tstar_space)
    Ustar_space = -cosine(eps, phi+0.5*np.pi, 1, tstar_space)
    return (Rstar_space, Ustar_space)

def linear_response_with_transient(tstar_space):
    #tstar = omega t;
    #assume non-dim parameters are set
    phi,eps = cosineparameters(pstar, wstar, lstar, omegastar, zetastar);
    _1 = np.ones(len(tstar_space));
    Rstar_space = _1 - cosine(eps, phi, 1, tstar_space) 
    #Add transient
    Rstar_space +=  np.exp(-zetastar*tstar_space)*cosine(eps, phi, np.sqrt(omegastar**2 - zetastar**2), tstar_space)
    Ustar_space = -cosine(eps, phi+0.5*np.pi, 1, tstar_space)
    return (Rstar_space, Ustar_space)
```

$$

$$
which gives

$$
\frac{d}{dt^*} 
\begin{bmatrix} R^* \\ U^* \end{bmatrix} =
\begin{bmatrix} U^* \\ \frac{1}{R^* }\left(

$$

## Integration of R-P (Non-dim)

```python
def pstar_infty (tstar):
    return lstar * (1 + pstar*np.cos(tstar))

#def dpinfty_dt(t):
#    return (-p_a*(omega*np.cos(omega*t) + (3*(t-t0)/t0**2)*np.sin(omega*t))*np.exp(-1.5*(t-t0)**2/(t0**2)));

#partial pressure of air as a function of R
def pstar_gas(Rstar, k):
    return lstar*(1+1/wstar)*np.power(Rstar, -3*k)

#d/dt (x) = equation(x,t)
def RP(x, tstar):
    (Rstar,Ustar) = x;
    dfdt = ( pstar_gas(Rstar, k) - pstar_infty(tstar) - 1.5*Ustar**2 
                     - 2*(zetastar)*Ustar/Rstar-lstar/(wstar*Rstar) )/Rstar
    return (Ustar, dfdt)

def RP_response(tstar_space):
    R_RP = odeint(RP, [1, 0], tstar_space)
    Ustar_space = R_RP[:,1]
    Rstar_space = R_RP[:,0]
    return Rstar_space, Ustar_space
```

## Integration of R-P (Original)
$$
\frac{d}{dt} 
\begin{bmatrix} R\\ U \end{bmatrix} =
\begin{bmatrix} U \\ \frac{1}{R }\left( 
     \frac{1}{\rho_L}(p_{\infty,0} +\frac{2\sigma}{R }) \left( \frac{R_0}{R}\right)^{3k} - \frac{1}{\rho_L} \left( p_{\infty,0} + p_A cos(\omega t) \right) - 
    \frac{3}{2} {U}^2 - \frac{4\mu}{\rho_L R_0} \frac{U}{R} - \frac{1}{\rho_L}\frac{2\sigma}{\rho}
\right)
\end{bmatrix}
$$

```python
def p_infty (t):
    p_it = p_atm * ( 1 +  pstar * np.cos(omega0*t/omegastar) )
    return p_it

#partial pressure of air as a function of R
def p_gas(R, k):
    p_g = p_atm*(1+1/wstar)*np.power(R_0/R, 3*k);
    return p_g

#d/dt (x) = equation(x,t)
def RPdim(x, t):
    (R,u) = x;
    udot = (p_gas(R, k) - p_infty(t)-2*sigma/R-4*mu*u/R)/(R*rho)-3*u**2/(2*R)
    return (u, udot)


def RPdim_response(R_0, U_0, t_space):
    R_RP = odeint(RPdim, [R_0, U_0], t_space)
    U_space = R_RP[:,1]
    R_space = R_RP[:,0]
    return R_space, U_space
```

## Frequency response (Linear approximation)

Amplitude of  (non-dimensionlised) radial perturbation is given by
$$
\epsilon_0 = 
  \frac{p^* l^*}{\sqrt{\left(  {(1 - {\omega^*}^2)}^2 + { ( 2\zeta^* )}^2        \right)}}
$$
And the maximum happens for the resonance frequency 
$$
    \omega_R = \omega_0 \sqrt{ 1 - 2{\zeta}^2 }    \quad\text{where}\quad  \zeta = \frac{\zeta^*}{\omega^*}   =  \frac{2\mu_L}{\rho_L R_0^2 \omega_0}
$$
and $max ~ (\epsilon) $ is 
$$
\epsilon_0 \bigg|_{\omega = \omega_R} =    \frac{p_A}{4\mu} \frac{1}{\sqrt{\omega_p^2   + \frac{1}{4} \left( \frac{4\mu}{\rho_L R_0^2}  \right)^2}}   = \frac{p_A}{4\mu} \frac{1}{\sqrt{\omega_0^2   -  \frac{1}{4} \left( \frac{4\mu}{\rho_L R_0^2}  \right)^2}}
$$

## Two limits ($\omega_0/\omega \ll 1$) and   ($\omega_0/\omega \gg 1$)

Amplitude of  (non-dimensionlised) radial perturbation is given by
$$
\epsilon_0 = 
  \frac{p^* l^*}{\sqrt{\left(  {(1 - {\omega^*}^2)}^2 + { ( 2\zeta^* )}^2        \right)}}
$$

### Case 1:  $\omega^* \ll 1$ or $\omega_0 \ll \omega$
$$
\epsilon_0 \bigg|_{\omega^* <<1} =  
  {p^* l^*} = \frac{p_A}{\rho_L R_0^2 \omega^2}
$$

### Case 2:  $\omega^* \gg 1$ or $\omega_0 \gg \omega$
$$
\epsilon_0 \bigg|_{\omega^* >> 1} = 
  \frac{p^* l^*}{{\omega^*}^2} =  \frac{p_A}{\rho_L R_0^2 \omega_0^2}
$$
