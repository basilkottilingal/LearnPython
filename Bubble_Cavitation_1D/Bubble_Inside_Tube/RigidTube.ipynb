# Linearisation of R-P equation in bulk
A bubble $R_0$ initially at equilibrium with the ambient pressure $p_{\infty,0}$ has a
equilibrium bubble pressure $p_{B,0} = p_{G,0} + p_V (T_\infty)$. The bubble response
in the limit of $M\rightarrow 0$ is given by R-P equation
\begin{equation}
  R\ddot{R} + \frac{3}{2} \dot{R}^2 + \frac{4\mu}{\rho_L} \frac{\dot{R}}{R} + \frac{2\sigma}{\rho_L R} =
  \frac{1}{\rho_L} \left(p_B(t) - p_\infty (t)\right)
\end{equation}
In the absence of vapor pressure and the gas inside bubble follows polytropic gas law,
we have
\begin{equation}
  p_B (t) = p_{G,0} \left( \frac{R_0}{R} \right)^{3k}
\end{equation}
If the ambient pressure is perturbed as
\begin{equation}
  p_\infty (t) = p_{\infty,0} + p_A ~R e \{ e^{j\omega t}\}
\end{equation}
with $p_A>0, ~p_A \ll p_{\infty,0}$
produces a linearised pertubation in the bubble radius from its equilibrium radius of the form
\begin{equation}
  R (t) = R_0 \left(1 - \epsilon_0 ~R e \{ e^{j(\omega t +\phi)} \} \right)
\end{equation}
with $\epsilon_0>0, ~\epsilon_0 \ll 1$.

\begin{eqnarray*}
  R\ddot{R} &=& \omega^2 R_0^2 (1 - \epsilon_0 { e^{j(\omega t +\phi)} })
                             \epsilon_0 { e^{j(\omega t +\phi)} } \\
  \frac{3}{2} \dot{R}^2 &=& \frac{3}{2} R_0^2 \omega^2 \epsilon_0^2 { e^{2j(\omega t +\phi)} }  \\
  \frac{4\mu_L}{\rho_L} \frac{\dot{R}}{R} &=& -j\omega\epsilon_0 ~{ e^{j(\omega t +\phi)} }
                                  (1+\epsilon_0 ~{ e^{j(\omega t +\phi)} })\frac{4\mu_L}{\rho_L} \\
  \frac{2\sigma}{\rho_L R} &=& \frac{2\sigma}{\rho_L R_0}
                   (1+\epsilon_0 { e^{j(\omega t +\phi)} } + \epsilon_0^2 ~{ e^{2j(\omega t +\phi)} } )\\
  p_B(t) &=& p_{G,0} (1+3k\epsilon_0 ~{ e^{j(\omega t +\phi)} }
                    +\frac{3k(3k-1)}{2}\epsilon_0^2 ~{ e^{2j(\omega t +\phi)} }) \\
  p_\infty(t) &=& p_{\infty,0} + p_A  e^{j \omega t }
\end{eqnarray*}
which gives
\begin{eqnarray}
  \epsilon_0^2 ~{ e^{2j(\omega t +\phi)} } \left(
    -\omega^2 R_0^2  + \frac{3}{2} R_0^2 -j\omega \frac{4\mu_L}{\rho_L} + \frac{2\sigma}{\rho_L R_0}
    - \frac{p_{G,0}}{\rho_L} \frac{3k(3k-1)}{2}
  \right) + \nonumber\\
  \epsilon_0 ~{ e^{j(\omega t +\phi)} } \left(
    \omega^2 R_0^2 - j\omega \frac{4\mu_L}{\rho_L} + \frac{2\sigma}{\rho_L R_0} - \frac{3kp_{G,0}}{\rho_L}
  \right)
  =
  \frac{1}{\rho_L} \left( p_{G,0} - \frac{2\sigma}{ R_0} - p_{\infty,0}  \right) - \frac{p_A}{\rho_L} e^{j\omega t}
\end{eqnarray}

# Natural Frequency inside rigid  tube [Oguz and Prosperetti]

A bubble centered at the axis $(0,0)$ inside a rigid tube of length $L$ and radius $R_{v}$ and a negligible thickness $\Gamma_t = \{ (r,z)~|~ r=R, ~-L/2 \leq z \leq L/2\}$. The bubble of initial radius $R_0$ is driven by acoustic cavitation which makes the bubble interface, $\Gamma$, oscillates

## Natural frequency of bubble oscillating inside tube

The bubble pressure inside the bubble and the liquid pressure just outside the interface can be found by stress balance.
(neglecting viscosity)
$$
    p_B(\mathbf{x},t) = p_L (\mathbf{x},t) + \sigma \kappa(\mathbf{x},t) \quad \forall \mathbf{x} \in \Gamma 
$$
    Using the unsteady incompressible potential flow equation in liquid $\frac{p_L}{\rho} = \frac{p_\infty}{\rho} - \frac{\partial \phi } {\partial t} - \frac{1}{2} \left( \nabla \phi \right)^2$. Neglecting $ \frac{1}{2} \left( \nabla \phi \right)^2$ w.r.t $\frac{\partial \phi } {\partial t}$ we get
$$
    p_{B}=-\rho \frac{\partial\phi}{\partial t}  + p_{\infty}+\sigma \kappa 
$$


Averaging on the onterface
$$
    p_{B}=-\rho\langle \frac{d\phi}{d t} \rangle +p_{\infty}+\sigma\langle \kappa \rangle,
$$
Differentiating
$$
    \frac{d}{d t}p_{B}=-\rho\langle \frac{d^2\phi}{d t^2} \rangle + \sigma\frac{d}{d t} \langle \kappa \rangle,
$$
    


$$
    \frac{d p_{B} }{d t} = \frac{d p_{B} }{d V} \frac{d V }{d t} 
$$
$${\frac{d V}{d t}}=\int_{\cal S}\mathbf{u}\cdot\mathbf{n}\ d{\cal S}\equiv{\cal S}\biggl<{\frac{\partial\phi}{\partial n}}\biggr>
$$

#### Spherical bubble case
If initial shape is sphere
$$
    \left< \kappa \right> (t) = \frac{2}{R_0} \left[1  - \frac{1}{j \omega R_0} \left< \frac{\partial \phi}{\partial n} \right>   e^{j\omega t} \right]
$$
$$
    \frac{\partial }{\partial t} \left< \kappa \right> (t) = - \frac{2}{R_0^2} \left< \frac{\partial \phi}{\partial n} \right>   e^{j\omega t}
$$
$$
    \frac{d p_{B} }{d t} = \frac{d p_{B} }{d V} \frac{d V }{d t} 
$$
$${\frac{d V}{d t}}=\int_{\cal S}\mathbf{u}\cdot\mathbf{n}\ d{\cal S} = {\cal S}\biggl<{\frac{\partial\phi}{\partial n}}\biggr>
$$


$$
    \frac{d p_{B} }{d V} {\cal S}\biggl<{\frac{\partial\phi}{\partial n}}\biggr> e^{j\omega t} =  \rho \omega^2 \langle \phi \rangle  e^{j\omega t} + - \frac{2\sigma}{R_0^2} \left< \frac{\partial \phi}{\partial n} \right>   e^{j\omega t}
$$

So this unforced undamped system will look like 

$$
    \left[ \omega^2 - \omega_0^2 \right] \rho \left< \phi \right> e^{j\omega t} = 0
$$
where the natural frequency of the system is given by 
$$\omega^{2}=\frac{1}{\rho}\left(S\,\frac{d p_{B}}{d V}+\frac{2\,\sigma}{R_0^{2}}\right)\,\frac{1}{\left\langle\phi\right\rangle}\left\langle\frac{\partial\phi}{\partial n}\right\rangle
$$
or 
where the natural frequency of the system is given by 
$$\omega_0^{2}=-\frac{1}{\rho R_0^2}\left(S R_0 \,\frac{d p_{B}}{d V}+\frac{2\,\sigma}{R_0}\right)\,\frac{-R_0}{\left\langle\phi\right\rangle}\left\langle\frac{\partial\phi}{\partial n}\right\rangle
$$

$$\frac{B_{n}}{A_{0}}=\frac{2}{n\pi}\,\frac{1}{I_{1}(n\pi R/L)}\;\int_{0}^{L}\frac{R_{v,0} ~R_0}{{(R_{v,0}^2 + (z-d)^2)}^{3/2}} \,\sin\biggl(\frac{n\pi z}{L}\biggr)d z, \quad \forall ~n>0

Using the orthogonality condition

$$\frac{C_{n}}{A_{0}}=-\,\frac{2}{R_{v,0}^{2}J_{0}^{2}(\,\alpha_{n})}\,\int_{0}^{R_{v,0}}\frac{r  R_0}{\sqrt{r^{2}+(L-d)^{2}}}\,J_{0}\bigg(\,\alpha_{n}\,\frac{r}{R_{v,0}}\bigg)d r, \quad \forall ~n>0
$$

In the limit $R_0/R_{v,0} \rightarrow 0$ we can write the potential with the  interface velocity
$$\phi = -\dot{R}(t)\, \left[ \frac{R^2 (t)}{\sqrt{r^2 + (z-d)^2}} \right]$$

from which you can take $A_0$ as $\dot{R} R $. Or in effect yo don't need to find $A_0$ as is will be cancelled in $\left<\frac{\partial \phi}{\partial n} \right>/\left< \phi\right>$


```python
from scipy import special
import math
## Constants in velocity potential functions
nterms = 5;
nsteps = 1000; 
def phi_params(L, d):
    ## All the dimensions are assumed to be normalized with R_v
    ## All constants A[i], B[i], C[i], D[i] are normalized with A[0]
    A = np.zeros(nterms);
    A[0] = 1.;
    B = np.zeros(nterms);
    C = np.zeros(nterms);
    D = np.zeros(nterms);
    alpha = special.jn_zeros(1, nterms)
    for n in range(1):
        C[n] = 0;
        for j in range(nsteps):
            dr =  1/nsteps;
            r = j*dr;
            #trapezoidal summation
            C[n] -=  2 * r * (R_0) * dr / ((r**2 +  (L-d)**2)**0.5) 
            r += dr
            C[n] -=  2 * r * (R_0) * dr / ((r**2 +  (L-d)**2)**0.5)
        D[n] = 0;
        for j in range(nsteps):
            dr =  1/nsteps;
            r = j*dr;
            #trapezoidal summation
            D[n] -=  2 * r * (R_0) * dr / ((r**2 +  d**2)**0.5) 
            r += dr
            D[n] -=  2 * r * (R_0) * dr / ((r**2 +  d**2)**0.5) 
        C[n] /= 2.
        D[n] /= 2.
    for _i in range(nterms-1):
        n = _i + 1;
        B[n] = 0;
        for j in range(nsteps):
            dz =  L/nsteps;
            z = j*dz;
            #trapezoidal summation
            B[n] +=  (R_0) * (2/(n*np.pi)) * (1/special.i1(n*np.pi/L) ) * np.sin(n*np.pi*z/L) * dz/((1 + (z-d)**2)**1.5)
            z += dz
            B[n] +=  (R_0) * (2/(n*np.pi)) * (1/special.i1(n*np.pi/L) ) * np.sin(n*np.pi*z/L) * dz/((1 + (z-d)**2)**1.5)
        C[n] = 0;
        for j in range(nsteps):
            dr =  1/nsteps;
            r = j*dr;
            #trapezoidal summation
            C[n] -=  (R_0) * (2/(special.j0(alpha[n]))**2) * r * special.j0(alpha[n]*r) * dr / ((r**2 +  (L-d)**2)**0.5) 
            r += dr
            C[n] -=  (R_0) * (2/(special.j0(alpha[n]))**2) * r * special.j0(alpha[n]*r) * dr / ((r**2 +  (L-d)**2)**0.5) 
        D[n] = 0;
        for j in range(nsteps):
            dr =  1/nsteps;
            r = j*dr;
            #trapezoidal summation
            D[n] -=  (R_0) * (2/(special.j0(alpha[n]))**2) * r * special.j0(alpha[n]*r) * dr / ((r**2 +  d**2)**0.5) 
            r += dr
            D[n] -=  (R_0) * (2/(special.j0(alpha[n]))**2) * r * special.j0(alpha[n]*r) * dr / ((r**2 +  d**2)**0.5) 
        B[n] /= 2.;
        C[n] /= 2.
        D[n] /= 2.
    return (A,  B, C, D, alpha);

```

### Velocity $\mathbf{u} = \frac{\partial \phi}{\partial r} \mathbf{e}_r + \frac{\partial \phi}{\partial z} \mathbf{e}_z$

```python

def velocity(params, r, z, R_0, L, d):
    A, B, C, D, alpha = params
    phi = 0.;
    ur = 0.;
    uz = 0.;
    # limited to 0 (for A[n])
    _rho = np.sqrt(r**2 + (z-d)**2);
    _p, _dp = special.lpn(0, (z-d)/_rho );
    for n in range(1):
        common = A[n] * ((R_0/_rho)** (n+1) )
        phi += common * _p[n];
        ur -=   common * ( _p[n] * (n+1) * r /( _rho**2 ) + _dp[n] * r * (z-d) / (_rho**3) );
        uz -=   common * ( _p[n] * (n+1) * (z-d) /( _rho**2 ) - _dp[n] * r * r / (_rho**3) );
    for _i in range(nterms-1):
        n = _i+1;
        phi += B[n] * np.sin(n*np.pi*z/L) * special.i0(n*np.pi*r/L)
        ur += B[n] * np.sin(n*np.pi*z/L) * special.i1(n*np.pi*r/L) *(n*np.pi/L);
        uz += B[n] * np.cos(n*np.pi*z/L) * special.i0(n*np.pi*r/L) *(n*np.pi/L);
    phi += C[0]*z/L + D[0]*(L-z)/L
    ur += 0;
    uz += (C[0]-D[0])/L
    for _i in range(nterms-1):
        n = _i+1;
        phi += (C[n]*math.sinh(alpha[n]*z)/math.sinh(alpha[n]*L) + D[n]*math.sinh(alpha[n]*(L-z))/math.sinh(alpha[n]*L)) * special.j0(alpha[n]*r)
        ur -=  (C[n]*math.sinh(alpha[n]*z)/math.sinh(alpha[n]*L) + D[n]*math.sinh(alpha[n]*(L-z))/math.sinh(alpha[n]*L)) * alpha[n] *special.j1(alpha[n]*r)
        uz +=  (C[n]*math.cosh(alpha[n]*z)/math.sinh(alpha[n]*L) - D[n]*math.cosh(alpha[n]*(L-z))/math.sinh(alpha[n]*L)) * alpha[n] *special.j0(alpha[n]*r)
    return (ur, uz, phi)


def velocity_bulk(params, r, z, R_0, L, d):
    _rho = np.sqrt(r**2 + (z-d)**2);
    common = (R_0/_rho)
    phi = common;
    ur  =  -common * r /( _rho**2 );
    uz  = - common * (z-d) /( _rho**2 )
    return (ur, uz, phi)


```

In cylindrical coordinate
$$ 
\left< f \right> = \frac{ \int  r f(z,r)  ~dz } {\int_z  R(\theta)  ~dz}
$$
$$ 
\left< f \right> = \frac{ \int_{0}^{\pi}  2\pi R^2 sin \theta f(Rcos\theta, Rsin\theta)  ~d\theta } { \int_{0}^{\pi}  2\pi R^2 sin \theta   ~d\theta }
$$
For sphere
$$ 
\left< f \right> = \frac{ \int_{0}^{\pi}  2\pi R^2 sin \theta f(Rcos\theta, Rsin\theta)  ~d\theta } { 4\pi R ^2 }

```python
def _1_average(R_0):
    #params etc are set
    _1_av = 0.;
    dtheta =  np.pi/nsteps;
    for i in range(nsteps):
        #left end of trapezium summ
        theta = i*dtheta;
        _1_av += 1 * 2*np.pi * np.sin(theta) *R_0 **2 * dtheta 
        #right end of trapezium summ
        theta += dtheta;
        _1_av += 1 * 2*np.pi * np.sin(theta) *R_0 **2 * dtheta 
    _1_av /= 2.; # Trapezium summation
    return _1_av/(4*np.pi*R_0**2);

_1_average(1)
```

```python
def phi_average(params, R_0, L, d):
    #params etc are set
    A, B, C, D, alpha = params
    phi_av = 0.;
    dtheta =  np.pi/nsteps;
    for i in range(nsteps):
        #left end of trapezium summ
        theta = i*dtheta;
        z = d + R_0 * np.cos(theta); 
        r = R_0 * np.sin(theta);
        ur, uz, phi = velocity(params, r, z, R_0, L, d);
        phi_av += phi * 2*np.pi * np.sin(theta) *R_0 **2 * dtheta
        #right end of trapezium summ
        theta += dtheta;
        z = d + R_0 * np.cos(theta); 
        r = R_0 * np.sin(theta);
        ur, uz, phi = velocity(params, r, z, R_0, L, d);
        phi_av += phi  * 2*np.pi * np.sin(theta) *R_0 **2 * dtheta
    phi_av /= 2.; # Trapezium summation
    phi_av /= 4.*np.pi*R_0**2
    return phi_av;

def dphidn_average(params, R_0, L, d):
    #params etc are set
    A, B, C, D, alpha = params
    dphidn_av = 0.;
    dtheta =  np.pi/nsteps;
    for i in range(nsteps):
        theta = i*dtheta;
        z = d + R_0 * np.cos(theta); 
        r = R_0 * np.sin(theta);
        ur, uz, phi = velocity(params, r, z, R_0, L, d);
        dphidn_av += (uz*np.cos(theta) + ur*np.sin(theta))  * 2*np.pi * np.sin(theta) *R_0 **2 * dtheta
        theta += dtheta;
        z = d + R_0 * np.cos(theta); 
        r = R_0 * np.sin(theta);
        ur, uz, phi = velocity(params, r, z, R_0, L, d);
        dphidn_av += (uz*np.cos(theta) + ur*np.sin(theta))  * 2*np.pi * np.sin(theta) *R_0 **2 * dtheta
    dphidn_av /= 2.; # Trapezium summation
    dphidn_av /= 4.*np.pi*R_0**2
    return dphidn_av;
```
